```python
import numpy as np

def needleman_wunsch(seq_a, seq_b, match=1, mismatch=-1, gap=-2):
    """
    Implement Needleman-Wunsch algorithm for global sequence alignment
    
    Parameters:
    - seq_a: First sequence
    - seq_b: Second sequence
    - match: Score for matching characters
    - mismatch: Penalty for mismatching characters
    - gap: Penalty for gaps
    
    Returns:
    - Optimal alignment score
    - Aligned sequences
    """
    
    # Initialize dimensions
    m, n = len(seq_a), len(seq_b)
    
    # Create scoring matrix (F) and initialize
    F = np.zeros((m + 1, n + 1), dtype=int)
    
    # Traceback matrix to reconstruct alignment
    traceback = np.zeros((m + 1, n + 1), dtype=int)
    # 0: stop, 1: diagonal, 2: up, 3: left
    
    # Initialize first row and column with gap penalties
    # This is the key difference from Smith-Waterman
    for i in range(m + 1):
        F[i, 0] = gap * i
    for j in range(n + 1):
        F[0, j] = gap * j
    
    # Fill the scoring matrix
    print("\nFilling the scoring matrix using Needleman-Wunsch algorithm...")
    print("="*70)
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            # Calculate scores for three possible moves
            if seq_a[i-1] == seq_b[j-1]:
                diagonal_score = F[i-1, j-1] + match
            else:
                diagonal_score = F[i-1, j-1] + mismatch
            
            up_score = F[i-1, j] + gap
            left_score = F[i, j-1] + gap
            
            # Needleman-Wunsch: take maximum (can be negative)
            F[i, j] = max(diagonal_score, up_score, left_score)
            
            # Record traceback direction
            if F[i, j] == diagonal_score:
                traceback[i, j] = 1
            elif F[i, j] == up_score:
                traceback[i, j] = 2
            else:
                traceback[i, j] = 3
    
    # Print the scoring matrix
    print("\nScoring Matrix F:")
    print("="*70)
    print("     ", end="")
    print("  -  ", end="")
    for char in seq_b:
        print(f"  {char}  ", end="")
    print()
    
    for i in range(m + 1):
        if i == 0:
            print("  -  ", end="")
        else:
            print(f"  {seq_a[i-1]}  ", end="")
        
        for j in range(n + 1):
            print(f" {F[i, j]:2d}  ", end="")
        print()
    
    optimal_score = F[m, n]
    print("\n" + "="*70)
    print(f"Optimal Global Alignment Score: {optimal_score}")
    print("="*70)
    
    # Traceback from bottom-right to top-left
    aligned_a = ""
    aligned_b = ""
    alignment_string = ""
    
    i, j = m, n
    while i > 0 or j > 0:
        if i > 0 and j > 0 and traceback[i, j] == 1:  # diagonal
            aligned_a = seq_a[i-1] + aligned_a
            aligned_b = seq_b[j-1] + aligned_b
            if seq_a[i-1] == seq_b[j-1]:
                alignment_string = "|" + alignment_string
            else:
                alignment_string = "x" + alignment_string
            i -= 1
            j -= 1
        elif i > 0 and (j == 0 or traceback[i, j] == 2):  # up
            aligned_a = seq_a[i-1] + aligned_a
            aligned_b = "-" + aligned_b
            alignment_string = " " + alignment_string
            i -= 1
        else:  # left (j > 0)
            aligned_a = "-" + aligned_a
            aligned_b = seq_b[j-1] + aligned_b
            alignment_string = " " + alignment_string
            j -= 1
    
    return optimal_score, aligned_a, aligned_b, alignment_string, F


# Main execution
if __name__ == "__main__":
    # Define sequences from the problem
    sequence_a = "ATCGATCGATCGATCG"
    sequence_b = "ATCGATCG"
    
    print("="*70)
    print("Problem 1: Global Alignment using Needleman-Wunsch Algorithm")
    print("="*70)
    print(f"\nSequence A: {sequence_a}")
    print(f"Sequence B: {sequence_b}")
    print(f"\nLength of A: {len(sequence_a)}")
    print(f"Length of B: {len(sequence_b)}")
    
    # Perform Needleman-Wunsch alignment
    score, aligned_a, aligned_b, alignment_str, matrix = needleman_wunsch(
        sequence_a, 
        sequence_b,
        match=1,      # Score for match
        mismatch=-1,  # Penalty for mismatch
        gap=-2        # Penalty for gap
    )
    
    # Display results
    print("\n" + "="*70)
    print("OPTIMAL GLOBAL ALIGNMENT RESULTS")
    print("="*70)
    print(f"\nOptimal Alignment Score: {score}")
    print(f"\nAligned Sequences:")
    print(f"A: {aligned_a}")
    print(f"   {alignment_str}")
    print(f"B: {aligned_b}")
    
    # Calculate alignment statistics
    matches = alignment_str.count('|')
    mismatches = alignment_str.count('x')
    gaps = alignment_str.count(' ')
    alignment_length = len(aligned_a)
    
    print(f"\n\nAlignment Statistics:")
    print(f"- Alignment length: {alignment_length}")
    print(f"- Matches: {matches}")
    print(f"- Mismatches: {mismatches}")
    print(f"- Gaps: {gaps}")
    print(f"- Identity: {matches}/{alignment_length} = {matches/alignment_length*100:.1f}%")
    
    print("\n" + "="*70)
    print("ANALYSIS COMPLETE!")
    print("="*70)
    
    # Additional analysis
    print("\n\nInterpretation:")
    print("-" * 70)
    print("The Needleman-Wunsch algorithm performs GLOBAL alignment, meaning it")
    print("aligns the entire sequences from end to end. This is different from")
    print("Smith-Waterman (local alignment) which finds the best matching region.")
    print("\nIn this case:")
    print(f"- Sequence A has {len(sequence_a)} nucleotides")
    print(f"- Sequence B has {len(sequence_b)} nucleotides")
    print(f"- The algorithm introduces {gaps} gap(s) to create the best global alignment")
    print("-" * 70)
    
```
