```python
import numpy as np

def smith_waterman(seq_x, seq_y, match=2, mismatch=-1, gap=-1):
    """
    Implement Smith-Waterman algorithm for local sequence alignment
    
    Parameters:
    - seq_x: First sequence
    - seq_y: Second sequence
    - match: Score for matching characters
    - mismatch: Penalty for mismatching characters
    - gap: Penalty for gaps
    
    Returns:
    - Best alignment score
    - Aligned sequences
    """
    
    # Initialize dimensions
    m, n = len(seq_x), len(seq_y)
    
    # Create scoring matrix (H) and initialize with zeros
    H = np.zeros((m + 1, n + 1), dtype=int)
    
    # Traceback matrix to reconstruct alignment
    traceback = np.zeros((m + 1, n + 1), dtype=int)
    # 0: stop, 1: diagonal, 2: up, 3: left
    
    max_score = 0
    max_pos = (0, 0)
    
    # Fill the scoring matrix
    print("\nFilling the scoring matrix using Smith-Waterman algorithm...")
    print("="*70)
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            # Calculate scores for three possible moves
            if seq_x[i-1] == seq_y[j-1]:
                diagonal_score = H[i-1, j-1] + match
            else:
                diagonal_score = H[i-1, j-1] + mismatch
            
            up_score = H[i-1, j] + gap
            left_score = H[i, j-1] + gap
            
            # Smith-Waterman: score cannot be negative (local alignment)
            H[i, j] = max(0, diagonal_score, up_score, left_score)
            
            # Record traceback direction
            if H[i, j] == 0:
                traceback[i, j] = 0
            elif H[i, j] == diagonal_score:
                traceback[i, j] = 1
            elif H[i, j] == up_score:
                traceback[i, j] = 2
            else:
                traceback[i, j] = 3
            
            # Track maximum score position
            if H[i, j] > max_score:
                max_score = H[i, j]
                max_pos = (i, j)
    
    # Print the scoring matrix
    print("\nScoring Matrix H:")
    print("="*70)
    print("     ", end="")
    print("  -  ", end="")
    for char in seq_y:
        print(f"  {char}  ", end="")
    print()
    
    for i in range(m + 1):
        if i == 0:
            print("  -  ", end="")
        else:
            print(f"  {seq_x[i-1]}  ", end="")
        
        for j in range(n + 1):
            print(f" {H[i, j]:2d}  ", end="")
        print()
    
    print("\n" + "="*70)
    print(f"Maximum score: {max_score} at position {max_pos}")
    print("="*70)
    
    # Traceback to find the alignment
    aligned_x = ""
    aligned_y = ""
    alignment_string = ""
    
    i, j = max_pos
    while i > 0 and j > 0 and H[i, j] != 0:
        if traceback[i, j] == 1:  # diagonal
            aligned_x = seq_x[i-1] + aligned_x
            aligned_y = seq_y[j-1] + aligned_y
            if seq_x[i-1] == seq_y[j-1]:
                alignment_string = "|" + alignment_string
            else:
                alignment_string = " " + alignment_string
            i -= 1
            j -= 1
        elif traceback[i, j] == 2:  # up
            aligned_x = seq_x[i-1] + aligned_x
            aligned_y = "-" + aligned_y
            alignment_string = " " + alignment_string
            i -= 1
        else:  # left
            aligned_x = "-" + aligned_x
            aligned_y = seq_y[j-1] + aligned_y
            alignment_string = " " + alignment_string
            j -= 1
    
    return max_score, aligned_x, aligned_y, alignment_string, max_pos, H


# Main execution
if __name__ == "__main__":
    # Define sequences
    sequence_x = "MAGNIFICENT"
    sequence_y = "MAGNIFIC"
    
    print("="*70)
    print("Problem 2: Local Alignment using Smith-Waterman Algorithm")
    print("="*70)
    print(f"\nSequence X: {sequence_x}")
    print(f"Sequence Y: {sequence_y}")
    print(f"\nLength of X: {len(sequence_x)}")
    print(f"Length of Y: {len(sequence_y)}")
    
    # Perform Smith-Waterman alignment
    score, aligned_x, aligned_y, alignment_str, max_pos, matrix = smith_waterman(
        sequence_x, 
        sequence_y,
        match=2,      # Score for match
        mismatch=-1,  # Penalty for mismatch
        gap=-1        # Penalty for gap
    )
    
    # Display results
    print("\n" + "="*70)
    print("BEST LOCAL ALIGNMENT RESULTS")
    print("="*70)
    print(f"\nAlignment Score: {score}")
    print(f"\nAligned Sequences:")
    print(f"X: {aligned_x}")
    print(f"   {alignment_str}")
    print(f"Y: {aligned_y}")
    
    print(f"\n\nAlignment Details:")
    print(f"- Number of matching positions: {alignment_str.count('|')}")
    print(f"- Length of aligned region: {len(aligned_x)}")
    print(f"- Identity: {alignment_str.count('|')}/{len(aligned_x)} = {alignment_str.count('|')/len(aligned_x)*100:.1f}%")
    
    print("\n" + "="*70)
    print("ANALYSIS COMPLETE!")
    print("="*70)
```
